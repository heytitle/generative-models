```python
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, _),(x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
print(x_train.shape)
print(x_test.shape)
```

```python
plt.imshow(x_train[0].reshape((28, 28)))
```

```python
EPOCH = 20
DROPOUT = 0.5
BATCH_SIZE = 64
TOTAL_BATCH = int(np.ceil(x_train.shape[0]/BATCH_SIZE))
```

# Shallow AutoEncoder

```python
x_train.shape
```

```python
tf.reset_default_graph()

latent_dim = 2

x = tf.placeholder(tf.float32, [None, 784])
ph_batch_size = tf.placeholder(tf.int64)
ph_latent = tf.placeholder(tf.float32, [None, latent_dim])

train_dataset = tf.data.Dataset.from_tensor_slices((x)).repeat().batch(ph_batch_size).shuffle(1000)
test_dataset = tf.data.Dataset.from_tensor_slices((x)).batch(ph_batch_size)

iterator = tf.data.Iterator.from_structure(train_dataset.output_types, train_dataset.output_shapes)

train_init_op = iterator.make_initializer(train_dataset)
test_init_op = iterator.make_initializer(test_dataset)

test_latent_walk_dataset = tf.data.Dataset.from_tensor_slices((ph_latent)).batch(ph_batch_size)
iterator2 = tf.data.Iterator.from_structure(test_latent_walk_dataset.output_types, test_latent_walk_dataset.output_shapes)
test_latent_walk_init_op = iterator2.make_initializer(test_latent_walk_dataset)



features = iterator.get_next()
latent_features = iterator2.get_next()


encoder = tf.layers.dense(features, latent_dim, activation=tf.nn.relu)

def decoder(in1, reuse=False):
    return tf.layers.dense(in1, 784, activation=tf.sigmoid, name='decoder', reuse=reuse)

x_hat = decoder(encoder)
x_hat_2 = decoder(latent_features, reuse=True)


loss = tf.losses.mean_squared_error(x_hat, features)

train_op = tf.train.AdamOptimizer(0.0005).minimize(loss)

sess = tf.Session()

```

```python
sess.run(tf.global_variables_initializer())

sess.run(train_init_op, feed_dict={ x: x_train, ph_batch_size: BATCH_SIZE})

for i in range(EPOCH):
    loss_accu = []
    for j in range(TOTAL_BATCH):
        _, loss_value = sess.run([train_op, loss])
        loss_accu.append(loss_value)
    print('Epoch %d| loss=%.4f' % (i+1, np.mean(loss_accu)))

sess.run(test_init_op, feed_dict={x: x_test, ph_batch_size: x_test.shape[0]})

x_original, x_latent, x_reconstucted = sess.run([features, encoder, x_hat])

x_original = x_original.reshape((-1, 28, 28))
x_reconstucted = x_reconstucted.reshape((-1, 28, 28))
```

```python
def walk_on_latent(dims=[0,1], total_samples=20):
    x_model = np.array([0,0])
    x_latent_walk = np.repeat(x_model.reshape((1, -1)), np.power(total_samples, 2), axis=0)
    a =  np.linspace(np.min(x_latent[:, 0]), np.max(x_latent[:, 0]), total_samples)
    b =  np.linspace(np.min(x_latent[:, 1]), np.max(x_latent[:, 1]), total_samples)
    xx, yy = np.meshgrid(a, b)

    for dd in dims:
        x_latent_walk[:, dd] = x_model[dd] + xx.reshape(-1)

    sess.run(test_latent_walk_init_op, feed_dict={ph_latent: x_latent_walk, ph_batch_size: xx.shape[0]**2})

    x_latent_walk = sess.run(x_hat_2)
    x_latent_walk_3d = x_latent_walk.reshape((-1, 28, 28))

    plt.figure(figsize=(10, 10))
    for i in range(total_samples):
        for j in range(total_samples):
            count = i*total_samples + j 
            plt.subplot(total_samples, total_samples, count + 1)
            plt.imshow(x_latent_walk_3d[count, :, :], cmap='gray')
            plt.axis('off')
            
walk_on_latent()
```
